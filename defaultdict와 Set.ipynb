### defaultdict  - 구현하는 세 가지 방법이 있음

1) 코드로 if - else 문

```python
words = ['apple','bat', 'bar', 'atom', 'book']
by_letter = {}
for word in words :
    letter = word[0]
    if letter not in by_letter :
        by_letter[letter] = [word]   # [word]라는 리스트로 하기
    else :
        by_letter[letter].append(word)
```

```python
by_letter
```

2) setdefault라는 메소드

```python
byy_letter ={}
for word in words :
    letter = word[0]
    byy_letter.setdefault(letter,[]).append(word)
```

```python
byy_letter
```

3) defaultdict라는 클래스

```python
from collections import defaultdict
byyy_letter = defaultdict(list)
```

```python
for word in words :
    byyy_letter[word[0]].append(word)
```

```python
byyy_letter
```
