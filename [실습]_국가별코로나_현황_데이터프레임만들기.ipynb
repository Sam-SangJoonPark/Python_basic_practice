```python
import requests
from bs4 import BeautifulSoup
url = "http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=14&ncvContSeq=&contSeq=&board_id=&gubun="
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
all = soup.find("div" , {"class" : "data_table mgt16"})

th = all.find_all("th",{'scope': 'col'})
th_cont = all.find_all("th", {"scope" : 'row'})
td = all.find_all("td")
```

```python
list_con = []
dict_country = {}
for i in td :
    i = i.text
    i = i.replace('\t','')
    i = i.replace('\r','')
    i = i.replace('\n','')
    list_con.append(i)
    

gukga = list_con[:370:2]
samang = list_con[1::2]
for a in range(len(gukga)) :
    dict_country[gukga[a]] = [samang[a]]
    
    
    
```

```python
th_cont
```

```python
gukga
```

```python
all_corona_file = pd.DataFrame(dict_country)
```

```python
all_corona_file =all_corona_file.T
```

```python
all_corona_file.columns = ['환자발새 수 (사망)']
```

```python
all_corona_file
```

```python
dict_all2 = {}
asia = []
middle = []
america = []
europe=[]
osania = []
africa= []
etc = []
for i in range(0,len(gukga)) :
    if i >= 0 and i < 29  :
        asia.append(gukga[i])
        
        dict_all2[gukga[i]] = 'asia'
    
    elif i>= 1+29-1 and i< 1+29+18-1 : 
        middle.append(gukga[i])
        
        dict_all2[gukga[i]] = 'middle'
    
    elif i >= 1+29+18-1 and i < 30+18+35-1 :
        america.append(gukga[i])
        
        dict_all2[gukga[i]] = 'america'
    
    elif i >= 30+18+35-1 and i < 30+18+35+49-1  :
        europe.append(gukga[i])
        
        dict_all2[gukga[i]] = 'europe'
    
    elif i >= 30+18+35+49-1 and i < 30+18+35+49+4-1  :
        osania.append(gukga[i])
        
        dict_all2[gukga[i]] = 'osania'
    
    elif i >= 30+18+35+49+4-1 and i < 30+18+35+49+4+49-1  :
        africa.append(gukga[i])
        
        dict_all2[gukga[i]] = 'africa'
    
    else :
        etc.append(gukga[i])
        
        dict_all2[gukga[i]] = 'etc'
        
   
```
