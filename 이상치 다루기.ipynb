## 이상치 다루어보자

```python
import pandas as pd
import numpy as np
# 일단 데이터 생성
data = pd.DataFrame(np.random.randn(1000,4))
data
```

```python
data.describe()
```

```python
# column 2에서 이상치를 뽑을 예정
col = data[2]
col
```

```python
# 절댓값으로 이상치를 감지할 것 
np.abs(col)
```

```python
# 절댓값 3넘는 것들
np.abs(col) > 3
```

```python
col[np.abs(col) > 3]
```

```python
data[(np.abs(data)>3).any(1)]
```

### 참고 ###
- any는 iterable 통과시키면 되고, any(0)은 인덱스가 칼럼이 나오고 any(1)은 인덱스가 인덱스가 나옴
- 방향은 이해 ok . 0은 row들 방향이라 아래로 내려가면서 체크하는거라 칼람별로 체크임 1은 반대
