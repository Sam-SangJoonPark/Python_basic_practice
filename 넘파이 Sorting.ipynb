```python
# 이러면 작은거부터 큰것으로 오름차순으로 정리 된다!
large_arr.sort()
large_arr
```

```python
# in1d(x,y)는 검색해보면 알겠지만 첫 째 배열 x가 둘째 배열 y의 원소를 포함하는지 여부임
# 관련 기능들로는 intersect1d , union1d, setdiff1d 등이 있음.
values = np.array([6, 0, 0, 3, 2, 5, 6])
np.in1d(values, [2, 3, 6])
```

## array 관련 operations 
 - unique(x) : sorted unique elements x를 뽑아냄
 - intersect1d(x,y) : x와 y의 common elements를 뽑아냄
 - union1d(x,y) : sorted union of elements를 뽑아냄
 - in1d(x,y) : 불린값 뽑아냄, x가 y의 원소들 안에 있는지
 - setdiff1d(x,y) : set들을 difference즉 x들 중에서 y에 없는거 뽑아냄
 - setxor1d(x,y) : difference 두 개임. 양쪽으로. 즉 x에있는거,y에없는거. y에있고 x에없는거 

```python
x = np.array([1,1,2,3,4,5,6,7])
y = np.array([10,10,20,30,40,5,6,7])
```

```python
# 유니크
np.unique(x)
```

```python
# 인터섹트
np.intersect1d(x, y)
```

```python
# 유니온. 합쳐서 유니크한
np.union1d(x,y)
```

```python
# 불린값으로. 지금 이것들 해보면 알겠지만 x,y가 길이가 동일할 필요는 없음. 
np.in1d(x,y)
```

```python
# set diffrence즉 x와 y 차집합
np.setdiff1d(x,y)
```

```python
# 차집합 양쪽 방향으로
np.setxor1d(x,y)
```
