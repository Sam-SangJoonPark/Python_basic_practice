# Missing Value를 다뤄보자!

```python
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
data = pd.Series([1,NA,3.5,NA,7])
data
```

```python
# 빈값 빼고 불러오기 
data[data.notnull()]
```

```python
data = pd.DataFrame([[1.,6.5,3.], [1.,NA,NA],
                   [NA,NA,NA], [NA,6.5,3.]])
data
```

```python
cleaned = data.dropna() # how = 'any'인 것임
```

```python
cleaned
```

```python
# 모두 NA인 것만 drop하고 싶다면 ?
data.dropna(how='all')
```

```python
# 이젠 열방향으로. 방향 바꿔서 모두가 NA인거 빼보자 
```

```python
data[4] = NA
data
```

```python
# axis=1 주면 열방향! 즉 열들이 나열된 방향쪽으로. 오른쪽으로.
data.dropna(axis=1, how='all')
```

## 결측치가 몇 개 이상이면 지우겠다는 조건 주기 

```python
df = pd.DataFrame(np.random.randn(7,3))
df
```

```python
df.iloc[:4,1] = NA
df.iloc[:2,2] = NA
df
```

```python
# df.dropna(thresh=2) 값이 두 개 이상 있는건 다 남긴다는거
df.dropna(thresh=1)   # 행에서 다 값이 적어도 1개는 있으니까 모두 남음
```

```python
df.dropna(thresh=2) # 두 개 이상인거 다 남음.
```

```python
df.dropna()
```

# 다른 값들로 채우려면 ?


```python
df.fillna('MISSING_VALUE')
```

# fillna 조건 좀 더 줘보기

```python
df.fillna({1:'ho', 2:'two'})   # 1과 2는 column명!
```

```python
df
```

# 참고 
 - fillna는 new object를 return해줌. 근데 원래 object를 바꾸려면 in-place 해주어야한다 

```python
df.fillna('change_new', inplace =False)
```

```python
df
```

```python
df.fillna('change_new', inplace=True)
```

```python
df
```

# 채우는 매소드!! 
 - ffill : 그냥 forward즉 앞에꺼 그냥 그대로 
 - bfill : back. 뒤꺼

```python
df = pd.DataFrame(np.random.randn(6,3))
df
```

```python
df.iloc[2:,1] =NA
df.iloc[4:,2] =NA
df
```

```python
# 그냥 앞에꺼 그대로 !
df.fillna(method='ffill') 
```

```python
df.fillna(method='ffill', limit=2) # 두개만 그렇게 하라는거
```

```python
# bfill 은 back 그대로 
df.fillna(method='bfill')
```

```python
df.loc[4,1] = 100
```

```python
df
```

```python
df.fillna(method = 'bfill')
```
